```python
import requests
import bs4
```

```python
site = requests.get('https://quotes.toscrape.com/')
```

```python
site
```

```python
site.text
```

```python
# Reading thorugh the page through Beautiful soup Module for HTML and Requests module for API call

readHtml = bs4.BeautifulSoup(site.text,'lxml')
readHtml
```

```python
# To get unique authors from the site pages

authors = set()

for author in readHtml.select('.author'):
    authors.add(author.text)
```

```python
authors
```

```python
# To get unique authors form all the pages, the scraping should stop once the page reaches last page and no authors

page_still_valid = True

authors = set()
page = 1

while page_still_valid:

    # Concatenate to get new page URL
    page_url = 'https://quotes.toscrape.com/page/{}/'.format(page)

    res = requests.get(page_url)

    if 'No quotes found!' in res.text:
        break

    # Turn into Soup
    soup = bs4.BeautifulSoup(res.text,'lxml')

    # Add Authors to our set
    for name in soup.select('.author'):
        authors.add(name.text)

    page = page +1
```
