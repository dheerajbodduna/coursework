```python
df.withColumn("Year",year(df['Date'])).show(3)
```

```python
#To show the average closing price per year

newdf = df.withColumn("Year",year(df['Date']))
```

```python
result = newdf.groupBy("Year").avg("Close")
```

```python
result.select("Year",format_number("avg(Close)",2).alias("Avg Close")).orderBy("Year").show()
```
