```python
# Break, Continue, Pass statements in Loops

# Break - Breaks out the current closest enclosing loop
x1 = 0

while x1<5:
  if x1 == 2:
    break
  print(x1)
  x1 = x1 +1
```

```python
# Continue - Goes to the top of the closest enclosing loop

x2 = 0

while x2 < 5:
    x2 = x2 + 1  # Increment x2 FIRST
    if x2 == 2:
        continue
    # Do other operations if needed.
    print(x2)

print('')

print(x2)
```

```python
# Pass - Does nothing

x3 = 0

while x3<5:
  if x3 == 2:
    pass
  x3 +=1
print('End of Loop')
```

Write a program that prints the integers from 1 to 100. But for multiples of three print "Fizz" instead of the number, and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".





```python
for i in range(1,100):
    if i%3==0 and i%5==0:
        print('FizzBuzz')
    elif i%5==0:
        print('Buzz')
    elif i%3==0:
        print('Fizz')
    else:
        print(i)
```

```python
#Function - Functions are standalone blocks of code that are defined independently of any class or object.

def greet(name):
    """This is a function."""
    return "Hello, " + name + "!"

message = greet("Alice")
print(message)  # Output: Hello, Alice!

#Methods - Methods are functions that are bound to objects (instances of classes).

class Dog:
    """This is a class."""
    def __init__(self, name):
        self.name = name

    def bark(self):
        """This is a method."""
        print("Woof! My name is " + self.name)

my_dog = Dog("Buddy")  # Create an object (instance) of the Dog class
my_dog.bark()        # Call the bark method on the my_dog object
# Output: Woof! My name is Buddy

```

```python
import random

num = random.randint(1,10)
#2

print("WELCOME TO GUESS ME!")
print("I'm thinking of a number between 1 and 10")
print("If your guess is more than 2 steps away from my number, I'll tell you you're Far!")
print("If your guess is within 2 steps of my number, I'll tell you you're Near!")
print("If your guess is farther than your most recent guess, I'll say you're getting Farther!")
print("If your guess is closer than your most recent guess, I'll say you're getting Nearer!")
print("LET'S PLAY!")

guesses = [0]

while True:

    # we can copy the code from above to take an input
    guess = int(input("I'm thinking of a number between 1 and 10.\n  What is your guess?"))
    #5

    if guess < 1 or guess > 10: # 5<1
        print('OUT OF BOUNDS! Please try again: ')
        continue

    # here we compare the player's guess to our number
    if guess == num:
        print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!!')
        break

    # if guess is incorrect, add guess to the list
    guesses.append(guess)

    # when testing the first guess, guesses[-2]==0, which evaluates to False
    # and brings us down to the second section

    if guesses[-2]:
        if abs(num-guess) < abs(num-guesses[-2]):
            print('Getting Nearer!')
        else:
            print('Getting Farther!')

    else:
        if abs(num-guess) <= 2:
            print('Near! Less than 2 steps!')
        else:
            print('Far! Greater than 2 steps!')
```

```python
work_hours = [('Arun',100),('Raju',200),('Vikram',300)]

def employee_check(work_hours):
  current_max = 0
  employee_of_month = ''

  for employee,hours in work_hours:
    if hours > current_max:
      current_max = hours
      employee_of_month = employee
    else:
      pass
  return (employee_of_month,current_max)
```

```python
#yield generator :without storing values in a variable

def gen_fibon(n):
  a = 1
  b = 1
  for i in range(n):
    yield a
    a,b = b,a+b
```

```python
list(gen_fibon(10))
```

```python
#Calling the variable :with storing values in a variable

def gen_fibon(n):
  a = 1
  b = 1
  output = []
  for i in range(n):
    output.append(a)
    a,b = b,a+b
  return output
```

```python
list(gen_fibon(10))
```
