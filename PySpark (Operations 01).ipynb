```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Ops').getOrCreate()
```

```python
df = spark.read.csv('/content/appl_stock.csv', inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.head(3)[0]
```

```python
df.filter("Close < 500").select(['Open', 'Close']).show()
```

```python
#Sql commands inside filter

df.filter("Close < 500").select(['Open', 'Close']).show(3)
```

```python
#Python commands inside filter

df.filter(df['Close']<500).select(['Open', 'Close']).show(3)
```

```python
#df Dataframe with Filter conditions like df[column]>x and not equals df[column]<x then show as dataframe frame top 3 rows

df.filter( (df['close']>200) & ~ (df['open']<200) ).show(3)
```

```python
#df Dataframe with Filter conditions like df[column]>x and not equals df[column]<x
#then collecting list top 1 row as Dictionary and volume key value

df.filter( (df['close']>200) & ~ (df['open']<200) ).collect()[0].asDict()['Volume']
```
