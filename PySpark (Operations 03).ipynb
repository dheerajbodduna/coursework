```python
# Importing SparkSession from pyspark.sql
from pyspark.sql import SparkSession
```

```python
## Initializing the spark with a sessionName

spark = SparkSession.builder.appName("Miss").getOrCreate()
```

```python
### Create a DataFrame from the local or cloud location format
df = spark.read.csv('/content/ContainsNull.csv',header=True,inferSchema=True)
```

```python
df.show()
```

```python
#Drops all the null values
df.na.drop().show()
```

```python
#Row needs to have 2 non-null values to show and the rest is dropped
df.na.drop(thresh=2).show()
```

```python
# how='any' -- if any of the value in a row is null then dropped
# how='all' -- if all of the values in a row is null then dropped

df.na.drop(how='all').show()
```

```python
#checks the subset column of the dataframe and consider specific column to consider and drop
df.na.drop(subset=['Sales']).show()
```

```python
#Fill null values

df.na.fill('No Name',subset=['Name']).show()
```

```python
from pyspark.sql.functions import mean
```

```python
mean_val = df.select(mean(df['Sales'])).collect()
```

```python
mean_sales = mean_val[0][0]
```

```python
df.na.fill(mean_sales,['Sales']).show()
```

```python
#Writing all aggs in one liner

df.na.fill(df.select(mean(df['Sales'])).collect()[0][0],['Sales']).show()
```
